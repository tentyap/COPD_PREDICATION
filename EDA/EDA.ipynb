# COPD in Nepal
### Factor that lead to COPD

1. Biomass Fuel Exposer (Indoor Air pollution )
2. Tobacco Smoking
3. Outdoor Air Pollution
4. Occupational Exposer
5. Genetic susceptability
6. Resperatory Infections 
7. Low Awerness and Health Acess
8. Climate and Enviromental Factors 



# sign and syntoms of copd
1. Chornic Cough
2. Shortness of Breath(dyspnea)
3. Excess mucus production(Sptum) (Phlega)
4. wheezing
5. Fatigue and low Energy
6. Chest Tightness 
7. Frequent resperatory Infections
8. Cynosis(Bluish lips or fingermail beds)
9. unintented Weight loss
10. Chornic Bronchitis and Empyhysenes systoms
11. Frequent Flare-ups (Exacerbations)


```python
# load libary
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Load a dataset
path = r"C:\Users\kamal\OneDrive\Desktop\COPD_PREDICATION\COPD_PREDICATION\Data\synthetic_COPD_data.csv"

df = pd.read_csv(path)

df.head()
```

```python
# set up a large  plot for better visilibity
plt.figure(figsize=(14,8))
```

```python
# get information about my data
df.info()
```

Univarate Anylisis

```python
# What is the Age Distrubation in my data
sns.histplot(df['Age'], kde=True, color='blue')
plt.title('Age Distrubation')
```

```python
# Bmi distrubation
sns.histplot(df['BMI'], kde=True, color='green')
plt.title('BMI Distrubation')
```

```python
# univarite analysis for the catogoraical features
# Gender distrubtion

sns.countplot(x = 'Gender', data=df, palette='viridis')
plt.title("Gender Distrubation")
```

```python
# Smoking Status(those who smoke and who never smoke)
sns.countplot(x = 'Smoking_Status', data=df, palette = 'Set1')
plt.title("Smoking status Distrubation")
```

# Bivariate Anylisis

```python
plt.figure(figsize=(14,10))
```

```python
# 1. Age vs Copd Digonesis
sns.boxplot(x= "COPD_Diagnosis",y='Age', data=df, palette='coolwarm')
plt.title('Age vs COPD Digonesis')
plt.show()
```

```python
# Bmi vs Copd Digonesis
sns.boxplot(x= "COPD_Diagnosis",y='BMI', data=df, palette='coolwarm')
plt.title("BMI vs COPD_digonesis")
```

```python
# Smoking status 
sns.boxplot(x = 'Smoking_Status', hue = 'COPD_Diagnosis', data=df , palette='Set2')
plt.title("smoking status vs COPD Digonesis")
plt.show()
```

```python
sns.countplot(x='COPD_Diagnosis', data=df, palette ='viridis')
plt.title("count plot of Digonis")
plt.show()
```

```python
# Correlation Matrix
data_corr = df[['Age',"Biomass_Fuel_Exposure",'Occupational_Exposure','Family_History_COPD','BMI','Air_Pollution_Level','Respiratory_Infections_Childhood','COPD_Diagnosis']]

corr = data_corr.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation matrix')
plt.show()
```
